## From Scratch
    * List of a list

```python
heading = ['States', 'Capital', 'LGAs', 'Population']
Info =[
    ['Lagos', 'Ikeja', 32, 245000],
    ['Kwara', 'Ilorin', 16, 223000],
    ['Osun', 'Osogbo', 27, 563455],
    ['Kano', 'Kano', 44, 798734]
]

data = pd.DataFrame(data=Info, columns=heading)
data
```

```python
stateData = {
    'States': ['Lagos', 'Kwara', 'Osun', 'Kano'],
    'Capital': ['Ikeja', 'Ilorin', 'Osogbo', 'Kano'],
    'LGAs': [32, 16,27,44],
    'Population': [ 245000, 223000, 563455, 798734],
}
```

```python
data = pd.DataFrame(data=stateData)
data
```
